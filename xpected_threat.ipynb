```python
events= events[['team','type','minute','location','pass_end_location','player','player_id']]
events = events[events['team']=='Chelsea'].reset_index()
```

```python
events = events.loc[(events['team']=='Chelsea') & (events['player']=='Frank Lampard')  & (events['type']=='Pass')  & (events['player_id']==20330.0)]
```

```python
print(events.to_string())
```

```python
events['location'] = events['location'].astype("string")
events['pass_end_location'] = events['pass_end_location'].astype("string")
```

```python
events.head(12)
```

```python
events[['x', 'y']] = events['location'].str.split(',', expand=True)
```

```python
events[['endx', 'endy']] = events['pass_end_location'].str.split(',', expand=True)
```

```python
events.head()
```

```python
events['x']= events['x'].str[1:]
events['y']= events['y'].str[:-1]
events['endx']= events['endx'].str[1:]
events['endy']= events['endy'].str[:-1]
```

```python
xT = pd.read_csv("/content/xT_Grid.csv", header=None)
xT = np.array(xT)
xT_rows, xT_cols = xT.shape
```

```python
xT # this is how our pitch is broken up
```

```python
events['x'] = pd.to_numeric(events['x'], errors='coerce')
events['y'] = pd.to_numeric(events['y'], errors='coerce')
events['endx'] = pd.to_numeric(events['endx'], errors='coerce')
events['endy'] = pd.to_numeric(events['endy'], errors='coerce')
```

```python
events['x1_bin'] = pd.cut(events['x'], bins=xT_cols, labels=False)
events['y1_bin'] = pd.cut(events['y'], bins=xT_rows, labels=False)
events['x2_bin'] = pd.cut(events['endx'], bins=xT_cols, labels=False)
events['y2_bin'] = pd.cut(events['endy'], bins=xT_rows, labels=False)
```

```python
events.head()
```

```python
events['start_zone_value'] = events[['x1_bin', 'y1_bin']].apply(lambda x: xT[x[1]][x[0]], axis=1)
events['end_zone_value'] = events[['x2_bin', 'y2_bin']].apply(lambda x: xT[x[1]][x[0]], axis=1)
```

```python
# now we find the difference between the start and end to get the xT value
events['xT'] = events['end_zone_value'] - events['start_zone_value']
```

```python
events.head(20)
```

```python
mean_xt = events['xT'].mean()

print(mean_xt)
```
